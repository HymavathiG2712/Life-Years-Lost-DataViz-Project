```python
import pandas as pd
import json

# Load the dataset
data_2015 = pd.read_excel("/content/drive/My Drive/Data Visualization/Data Visualization Project/2015_small.xlsx")

# Load the JSON file with code mappings
with open("/content/drive/My Drive/Data Visualization/Data Visualization Project/2015_codes.json", 'r') as file:
    codes_2015 = json.load(file)
```

```python
# Replace codes in the dataset for all columns with corresponding mappings in the JSON file
for column in data_2015.columns:
    if column in codes_2015:
        data_2015[column] = data_2015[column].astype(str).map(codes_2015[column]).fillna(data_2015[column])

# Display the transformed data for the 'education_2003_revision' column as a test
data_2015.head()
```
